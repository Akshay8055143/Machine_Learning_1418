## Separate X and Y features
Y : Species X : all remaining features, sepal length/width, petal length/width

```python
df.columns
```

```python
X = df.drop(columns=['price_range'])
Y = df['price_range']
```

```python
X.head()
```

```python
Y.head()
```

## Data Preprocessing and Data Cleaning

```python
from sklearn.pipeline import make_pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
```

```python
pre = make_pipeline(
    SimpleImputer(strategy='median'),
    StandardScaler()
).set_output(transform='pandas')
```

```python
X_pre = pre.fit_transform(X)
X_pre.head()
```

## Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(X_pre,Y,train_size=0.7,test_size=0.3,random_state=21)
```

```python
xtrain.head()
```

```python
xtrain.index
```

```python
xtest.head()
```

```python
ytrain.head()
```

```python
ytest.head()
```

## Build the Model

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
model.fit(xtrain,ytrain)
```

```python
model.score
```

```python
ypred_train = model.predict(xtrain)
```

```python
yprob_train = model.predict_proba(xtrain)
```

```python
yprob_train[0]
```

```python
ypred_train[:5]
```

```python
ytrain.head()
```

## Model Evaluation

```python
model.score(xtrain,ytrain)
```

```python
model.score(xtest,ytest)
```

## Confusion Matrix

```python
from sklearn.metrics import ConfusionMatrixDisplay
ConfusionMatrixDisplay.from_estimator(model,xtest,ytest)
```

## Classification Report

```python
ypred_test = model.predict(xtest)
```

```python
from sklearn.metrics import classification_report
print(classification_report(ytest,ypred_test))
```

```python
from sklearn.metrics import f1_score
f1_score(ytest,ypred_test,average='macro')
```

## All the scores are greater than 80%(0.80), we can consider this model for final model building
